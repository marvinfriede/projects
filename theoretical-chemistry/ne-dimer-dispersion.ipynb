```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', size=16)          # controls default text sizes
plt.rc('axes', titlesize=16)     # fontsize of the axes title
plt.rc('axes', labelsize=18)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=16)    # fontsize of the tick labels
plt.rc('ytick', labelsize=16)    # fontsize of the tick labels
plt.rc('legend', fontsize=18)    # legend fontsize
plt.rc('figure', titlesize=18)   # fontsize of the figure title
```

# Ne dimer: PES (BLYP, BLYP-D3BJ)

```python
no_disp = np.genfromtxt("data/ne-dimer-pes-nodisp.txt")
no_disp_dist = no_disp[:, 0]
no_disp_energy = no_disp[:, 1]
no_disp_min = no_disp_dist[np.argmin(no_disp_energy)]

d3bj = np.genfromtxt("data/ne-dimer-pes-disp.txt")
d3bj_dist = d3bj[:, 0]
d3bj_energy = d3bj[:, 1]
d3bj_min = d3bj_dist[np.argmin(d3bj_energy)]

f = plt.figure(figsize=(20,10))
plt.plot(no_disp_dist, no_disp_energy, color="r", marker=None, ls="-", label="no dispersion")
plt.plot(no_disp_min, np.amin(no_disp_energy), color="r", marker="o", label="{:.2F}".format(no_disp_min))
plt.plot(d3bj_dist, d3bj_energy, color="b", marker=None, ls="-", label="D3BJ")
plt.plot(d3bj_min, np.amin(d3bj_energy), color="b", marker="o", label="{:.2F}".format(d3bj_min))
plt.xlabel("R(Ne-Ne) in Angstrm")
plt.ylabel("E in a.u.")
plt.legend(loc="best")
plt.show()
```

# Ne dimer: PES (CCSD(T))

```python
cc = np.genfromtxt("data/ne-dimer-pes-ccsdt.txt")
cc_dist = cc[:, 0]
cc_energy = cc[:, 1]
cc_min = cc_dist[np.argmin(cc_energy)]

f = plt.figure(figsize=(20,10))
plt.plot(cc_dist, cc_energy, color="k", marker=None, ls="-", label="CCSD(T)")
plt.plot(cc_min, np.amin(cc_energy), color="k", marker="o", label="{:.2F}".format(cc_min))
plt.xlabel("R(Ne-Ne) in Angstrm")
plt.ylabel("E in a.u.")
plt.legend(loc="best")
plt.show()
```
