```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', size=16)          # controls default text sizes
plt.rc('axes', titlesize=16)     # fontsize of the axes title
plt.rc('axes', labelsize=18)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=16)    # fontsize of the tick labels
plt.rc('ytick', labelsize=16)    # fontsize of the tick labels
plt.rc('legend', fontsize=18)    # legend fontsize
plt.rc('figure', titlesize=18)   # fontsize of the figure title
```

# H<sub>2</sub> dissociation curve

```python
data = np.genfromtxt("data/h2-diss-rhf.dat")
data_mp2 = np.genfromtxt("data/h2-diss-mp2.dat")

f = plt.figure(figsize=(20,10))
plt.plot(data[:, 0], data[:, 1], color="r", marker="", ls="-", label="RHF")
plt.plot(data_mp2[:, 0], data_mp2[:, 1], color="b", marker="", ls="-", label="MP2")
plt.axhline(y=-0.4798356*2, color="k", ls="--", label="Dissociation limit RHF")
plt.xlabel("R(H-H) in Bohr")
plt.ylabel("E in a.u.")
plt.legend(loc="best", fancybox=True, shadow=True)
plt.show()
```

# He<sub>2</sub> dissociation curve

```python
data_he_atom = np.genfromtxt("data/he-atom.dat")
data_he_atom_mp2 = np.genfromtxt("data/he-atom-mp2.dat")
data_he_atom_cp = np.genfromtxt("data/he-atom-ghost.dat")
data_he_dimer = np.genfromtxt("data/he2-diss-rhf.dat")
data_he_dimer_mp2 = np.genfromtxt("data/he2-diss-mp2.dat")

# distance, only once needed
dist = data_he_atom[:, 0]

# relative to lowest value
atom_energy = data_he_atom[:, 1]
atom_energy_mp2 = data_he_atom_mp2[:, 1]
atom_cp_energy = data_he_atom_cp[:, 1]
dimer_energy = data_he_dimer[:, 1]
dimer_energy_mp2 = data_he_dimer_mp2[:, 1]

# get dist with lowest energy
rhf_min_energy = dimer_energy - 2*atom_energy
rhf_min = dist[np.argmin(rhf_min_energy)]
mp2_min_energy = dimer_energy_mp2 - 2*atom_energy_mp2
mp2_min = dist[np.argmin(mp2_min_energy)]

f = plt.figure(figsize=(20,10))
plt.plot(dist, dimer_energy - 2*atom_cp_energy, color="r", marker="", ls="-", label="RHF (CP)")
plt.plot(dist, dimer_energy - 2*atom_energy, color="b", marker="", ls="-", label="RHF (no CP)")
plt.plot(rhf_min, np.amin(rhf_min_energy), color="b", marker="o", label="{:.2F} Bohr".format(rhf_min))
plt.plot(dist, dimer_energy_mp2 - 2*atom_energy_mp2, color="g", marker="", ls="-", label="MP2")
plt.plot(mp2_min, np.amin(mp2_min_energy), color="g", marker="o", label="{:.2F} Bohr".format(mp2_min))he-atom-mp2
plt.xlabel("R(H-H) in Bohr")
plt.ylabel("E in a.u.")
plt.xlim(5,10)
plt.ylim(-0.00007, 0.0001)
plt.legend(loc="best", fancybox=True, shadow=True)
plt.show()
```

# Charge density

```python
data_h2 = np.genfromtxt("data/density_h2.dat")
data_he = np.genfromtxt("data/density_he.dat")
data_be = np.genfromtxt("data/density_be.dat")

f = plt.figure(figsize=(20,10))
plt.plot(data_h2[:, 0], data_h2[:, 1], color="r", marker="", ls="-", label="H2 (bond)")
plt.plot(data_he[:, 0], data_he[:, 1], color="b", marker="", ls="-", label="He")
plt.plot(data_be[:, 0], data_be[:, 1], color="g", marker="", ls="-", label="Be")
plt.xlabel("z-coordinate")
plt.ylabel("charge density")
plt.legend(loc="best")
plt.show()
```
