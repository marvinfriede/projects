```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', size=16)          # controls default text sizes
plt.rc('axes', titlesize=16)     # fontsize of the axes title
plt.rc('axes', labelsize=18)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=16)    # fontsize of the tick labels
plt.rc('ytick', labelsize=16)    # fontsize of the tick labels
plt.rc('legend', fontsize=18)    # legend fontsize
plt.rc('figure', titlesize=18)   # fontsize of the figure title
```

# H<sub>2</sub> dissociation

```python
data_rhf = np.genfromtxt("data/h2-pes-rhf.trjact.dat")
data_uhf = np.genfromtxt("data/h2-pes-uhf.trjact.dat")
data_mp2 = np.genfromtxt("data/h2-pes-mp2.trjact.dat")
data_ccsd = np.genfromtxt("data/h2-pes-ccsd.trjact.dat")

f = plt.figure(figsize=(20,10))
plt.plot(data_rhf[:, 0], data_rhf[:, 1], color="r", marker="", ls="-", label="RHF")
plt.plot(data_uhf[:, 0], data_uhf[:, 1], color="b", marker="", ls="-", label="UHF")
plt.plot(data_mp2[:, 0], data_mp2[:, 1], color="g", marker="", ls="-", label="MP2")
plt.plot(data_ccsd[:, 0], data_ccsd[:, 1], color="k", marker="", ls="-", label="CCSD")
plt.plot()
plt.xlabel("R(H-H) in Angström")
plt.ylabel("E in a.u.")
plt.ylim(-1.2, -0.7)
plt.xlim(0, 6)
#plt.xscale("log")
plt.legend(loc="best")
plt.show()
```
