```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial # allows np.array

plt.rc('font', size=16)          # controls default text sizes
plt.rc('axes', titlesize=16)     # fontsize of the axes title
plt.rc('axes', labelsize=18)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=16)    # fontsize of the tick labels
plt.rc('ytick', labelsize=16)    # fontsize of the tick labels
plt.rc('legend', fontsize=18)    # legend fontsize
plt.rc('figure', titlesize=18)   # fontsize of the figure title
```

```python
def createRange(start, stop, step):
    return np.linspace(start, stop, int((stop - start) / step + 1))
```

```python
x = createRange(0, 1e4, 1)
xFactorial = createRange(0, 30, 0.1)

f = plt.figure(figsize=(20,10))
plt.plot(x, x**4, color="r", marker="", ls="-", label=r"$N^{4}$ (HF)")
plt.plot(x, x**5, color="g", marker="", ls="-", label=r"$N^{5}$ (MP2)")
plt.plot(x, x**6, color="#ffa500", marker="", ls="-", label=r"$N^{6}$ (CCSD)")
plt.plot(x, x**7, color="b", marker="", ls="-", label=r"$N^{7}$ (CCSD(T))")
plt.plot(x, x**8, color="y", marker="", ls="-", label=r"$N^{8}$ (CCSDT)")
plt.plot(xFactorial, factorial(xFactorial), color="k", marker="", ls="-", label=r"$N!$ (FCI)")
plt.yscale('log')
plt.xlabel("$N$ (number of basis functions)")
plt.ylabel("'time equivalent'")
plt.title("Visualization of the scaling behaviour of quantum chemical methods.")
plt.legend(loc="best")
plt.show()
```
